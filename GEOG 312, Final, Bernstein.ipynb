```python
# 1. Data Manipulation (5 points): Write a Python function that takes a list of lists representing 
# points (latitude, longitude) and returns the average latitude and average longitude

mount_everest = [27.9880556, 86.9252778]
makalu = [27.8897222, 87.0888889]
manaslu = [28.55, 84.5597222]
lhotse = [27.9616667, 86.9330556]

mountains = [mount_everest, makalu, manaslu, lhotse]

mount_number = 0
latitude_sum = 0.0
longitude_sum = 0.0

while mount_number < 4:
    latitude_sum = mountains[mount_number][0] + latitude_sum 
    longitude_sum = mountains[mount_number][1] + longitude_sum 
    mount_number += 1

latitude_av = latitude_sum / 4
longitude_av = longitude_sum / 4

print("Latitude Average:", latitude_av)
print("Longitude Average:", longitude_av)
```

```python
# Develop a Python program that automates a repetitive GIS task you encountered during your project or coursework. 
# Explain the task, your approach, and the benefits of automation.

!pip install geopandas
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

def csv_to_shapefile():
    csv_file_name = input("What is yout file name? Ensure that it is correct and ends in '.csv' please. ")
    Latitude = input("Input the title of your Latitude Column: ")
    Longitude = input("Input the title of your Longitude Column: ")
    CRS = input("Input your desired crs: ")
    Output_name = input("Name your Shapefile output title. Remember to end it with '.shp' :")
    stepOne = pd.read_csv(csv_file_name)
    stepTwo = [Point(xy) for xy in zip(stepOne[Latitude], stepOne[Longitude])]
    stepThree = gpd.GeoDataFrame(stepOne, geometry = stepTwo, crs = CRS)
    output_shapefile = Output_name
    stepThree.to_file(output_shapefile)
    print("Your csv is now shapefile under the name: ", Output_name)

csv_to_shapefile()
    
# This code turns a CSV file into a Shapefile.
# My approach was to create a function that in-took all neccessary information, and easily provided a shapefile for the user.
# The benefits of this is so the user does not have to create their own code each time, and they don't risk errors.
```
